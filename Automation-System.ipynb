```python
from abc import ABC, abstractmethod
class BuildingSystem(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass
    @abstractmethod
    def status(self):
        pass

class AirConditioningSystem(BuildingSystem):
    def start(self):
        print("Started AirConditioningSystem")

    def stop(self):
        print("Stopped AirConditioningSystem")

    def status(self):
        print("Status: Cooling the Room")

class LightSystem(BuildingSystem):
    def start(self):
        print("Lighting system turned on")

    def stop(self):
        print("Lighting system turned off")

    def status(self):
        print("Lighting system turned on")

class SecuritySystem(BuildingSystem):
    def start(self):
        print("Security system activated")

    def stop(self):
        print("Security system deactivated")

    def status(self):
        print("Status: Checking the Security")

class FireAlarmSystem(BuildingSystem):
    def start(self):
        print("Fire system activated")

    def stop(self):
        print("Fire system deactivated")

    def status(self):
        print("Status: Checking the Fire outbreak")

system = [
    AirConditioningSystem(),
    LightSystem(),
    SecuritySystem(),
    FireAlarmSystem()
]
for system in system:
    system.start()
    system.stop()
    system.status()




```
